```python
import numpy as np


# S = {
#     "<" : {"x1": 1, "x2": 3, "b": 200},
#     "<" : {"x1": 1, "x2": 1, "b": 350},
#     "<" : {"x1": 1, "x2": 1, "b": 450},
#     "f" : {"x1": 1, "x2": 2}
# }


def Dual(A, b, c, con, prob):
    TA = np.transpose(A)
    C = b
    B = c

    for i in range(len(con)):
        if prob=="Max":
            con[i]=">"
        elif prob=="Min":
            con[i]=="<"
    if prob == "Max":
        prob = "Min"
    elif prob == "Min":
        prob = "Max"
    return TA, C, B, con, prob






prob = "Max"

A = np.array([[1, 3],
              [2, 1],
              [1, 1]])
b = np.array([450, 350, 200])
c = np.array([1, 2])
con = ["<", "<", "<"]

print(f"\n{prob} = ", end="")
for i in range(len(c)):
    if i==len(c)-1:
        print(f"{c[i]} x{i+1} ", end="")
    else:
        print(f"{c[i]} x{i + 1} + ", end="")

print(end="\n\n")
for i in range(len(A)):
    for j in range(len(A[i])):
        if j==len(A[i])-1:
            print(A[i][j], f"x{j+1}", " ", end="")
        else:
            print(A[i][j], f"x{j+1}", " + ", end="")
    print(con[i], "", b[i])



A, c, b, con, prob = Dual(A, b, c, con, prob)

print("\nDual : ")

print(f"\n{prob} = ", end="")
for i in range(len(c)):
    if i==len(c)-1:
        print(f"{c[i]} y{i+1} ", end="")
    else:
        print(f"{c[i]} y{i + 1} + ", end="")

print(end="\n\n")
for i in range(len(A)):
    for j in range(len(A[i])):
        if j==len(A[i])-1:
            print(A[i][j], f"y{j+1}", " ", end="")
        else:
            print(A[i][j], f"y{j+1}", " + ", end="")
    print(con[i], "", b[i])
```
