```python
import numpy as np
#pour tracer le graphe TSP
import matplotlib.pyplot as plt
import  random as rd



N=5 #le nombre de villes

T0=10
T=T0
Tmin=9
tau=1e4


beta=1e-1
k=0
t=0

def fonction_cout():
    global chemin
    energie =0.0
    #xy = np.concatenate((x[chemin],y[chemin]))
    xy=np.column_stack((x[chemin],y[chemin]))
    energie= np.sum(np.sqrt(np.sum((xy - np.roll(xy,-1, axis=0))**2,axis=1)))
    return energie




def Deplacement(ville1,ville2):
 

 
    global chemin
    #print('chemin courant',chemin)
    #print('ville1,ville2 choisies alétoirement',ville1,ville2)
    
    Min=min(ville1,ville2)
    Max=max(ville1,ville2)
    #print('min =',Min,'max =',Max)
    
    chemin[Min:Max]= chemin[Min:Max].copy()[::-1]
    #print('chemin [Min:Max]', chemin[Min:Max])
    #print('chemin [Min:Max].copy()[::-1]',chemin[Min:Max].copy()[::-1])
    #print('chemin nouveau',chemin)
    return

  
x = np.random.uniform(0, 1, N)
y = np.random.uniform(0, 1, N)

#choix du chemin initiale
chemin = np.arange(N)
chemin1 = chemin.copy()

cout_courant = fonction_cout()
d1=cout_courant
print('k=',k, 'Chemin Hamiltonien :',chemin,'distance du chemin =',cout_courant,'Température',T)


while  T > Tmin :  
    
   
    ville_voisine = rd.randint(1,N-1)  
    ville_courante = rd.randint(1,N-1)
    if ville_voisine == ville_courante:continue

    Deplacement(ville_voisine,ville_courante)
    
    cout_nouveau = fonction_cout()

    if cout_nouveau <= cout_courant:
        cout_courant= cout_nouveau
    else:  #Critère de Métropolis
        dE = cout_nouveau-cout_courant
        #beta=rd.uniform(0,1)
        if beta > np.exp(-dE/T):
            Deplacement(ville_voisine,ville_courante)  
        else:  
            cout_courant = cout_nouveau           

    t+=1
    T = T0*np.exp(-t/tau)
    k+=1 

    # Abaisser la température ======> Refroidissement
    print('k =', k, 'Chemin Hamiltonien : ',chemin,' Distance du chemin= ', cout_courant, 'Température',T)
    #plt.plot(x[chemin],y[chemin],marker='o',color='r')


print('Résultat final')
print('chemin initial:',chemin1,"Distance du chemin initial =",d1,'température initiale =',T0)
print('chemin optimal:',chemin,"Distance du chemin optimal =",cout_courant,'température Final =',T)
#plt.plot(x[chemin1],y[chemin1],marker='o',color='r')
plt.plot(x[chemin],y[chemin],marker='o',color='g')
```
